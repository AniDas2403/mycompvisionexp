#### NOTE: OpenCV imports image in BGR format

## Importing Libraries

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

## Inputting and printing test image

```python
img = cv2.imread('TestImg.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

#### Obtaining Img Dimensions

```python
H, W, C = img.shape
print(f"Image height: {H}")
print(f"Image width: {W}")
print(f"Number of image channels: {C}")
```

## Setting red channel to 0 for all pixels

```python
def red0(img):
    i = np.zeros(shape=[H, W, C], dtype=np.uint8)
    for h in range(H):
        for w in range(W):
            i[h][w] = img[h][w] * [1,1,0]
    
    return i
```

#### Testing function

```python
i1 = red0(img)
plt.imshow(cv2.cvtColor(i1, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.imshow(i1)
```

## Converting Img to Grayscale

```python
def gs_convert(img):
    i = np.zeros(shape=[H, W, C], dtype=np.uint8)
    for h in range(H):
        for w in range(W):
            i[h][w] = (img[h][w][0] * 0.114) + (img[h][w][1] * 0.587) + (img[h][w][2] * 0.299)
    
    return i
```

#### Testing function

```python
i2 = gs_convert(img)
plt.imshow(cv2.cvtColor(i2, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.imshow(i2)
```

## Shifting Color and Clamping

```python
def shift_color(img,RV):
    i = np.zeros(shape=[H, W, C], dtype=np.uint8)
    for h in range(H):
        for w in range(W):
            i[h][w][0] = img[h][w][0] + RV[0]
            if (i[h][w][0] > 255):
                i[h][w][0] = 255
            elif (i[h][w][0] < 0):
                i[h][w][0] = 0
            
            i[h][w][1] = img[h][w][1] + RV[1]
            if (i[h][w][1] > 255):
                i[h][w][1] = 255
            elif (i[h][w][1] < 0):
                i[h][w][1] = 0
                
            i[h][w][2] = img[h][w][2] + RV[2]
            if (i[h][w][2] > 255):
                i[h][w][2] = 255
            elif (i[h][w][2] < 0):
                i[h][w][2] = 0
    
    return i
```

#### Enter shift matrices for testing

```python
# Enter test vector in BGR format
RV1 = [7, 53, -27]
RV2 = [243, -210, 9]
```

#### Testing function

```python
i3a = shift_color(img,RV1)
plt.imshow(cv2.cvtColor(i3a, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.imshow(i3a)
```

```python
i3b = shift_color(img,RV2)
plt.imshow(cv2.cvtColor(i3b, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.imshow(i3b)
```
